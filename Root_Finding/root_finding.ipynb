#Bisection Method

```python
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
func = x**2 - 4
f = sp.lambdify(x, func)
a = 0
b = 3
tol = 1e-6
```

```python
def bisection(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("Functions must have opposite signs, quitting")

    while (b - a) / 2 > tol:
        midpoint = (a + b) / 2
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint

    return midpoint
```

```python
root = bisection(f, a, b, tol)
print(f"Root: {root}")
```
