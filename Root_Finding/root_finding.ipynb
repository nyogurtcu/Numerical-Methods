#Bisection Method

```python
import numpy as np
import sympy as sp
from sympy import sin, cos, exp, sqrt, log as ln
```

```python
x = sp.symbols('x')
func_f = x**3 - 2*x -5 #For Bisection Method
func_df = sp.diff(func_f)
func_g = cos(x) #For FPI
f = sp.lambdify(x, func_f)
g = sp.lambdify(x,func_g)
df = sp.lambdify(x, func_df)
tol = 1e-6
```

```python
a = 0
b = 4
def bisection(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("Functions must have opposite signs, quitting")

    while (b - a) / 2 > tol:
        midpoint = (a + b) / 2
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint

    return midpoint
```

```python
root = bisection(f, a, b, tol)
root
```

#Fixed Point Iteration

```python
def fixed_point_iter(g, x, tol, max_iter=10):
    x_values = np.zeros((max_iter +1,1))
    x_values[0] = 1
    for i in range(max_iter):
        x_values[i+1] = g(x_values[i])
        if abs(x_values[i+1] - x_values[i]) < tol:
            break
    return x_values
```

```python
root = fixed_point_iter(g, x, tol, 10)
```

#Newton's Method

```python
def newtons_method(f, df, x0, tol, max_iter):
    x = np.zeros((max_iter+1,1))
    x[0]=x0
    for i in range(max_iter):
        if df(x[i]) == 0:
            return None
        x[i+1] = x[i] - f(x[i])/df(x[i])

        error = x[i+1] - x[i]
        if abs(error) < tol:
            return x[:i+2]

    return x
```

```python
roots = newtons_method(f, df, 2, tol, 10)
```
